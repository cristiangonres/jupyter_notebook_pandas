```python
import random
def calcular(retweets):
    return retweets * random.randint(3, 5)

df["ganancias"] = df["retweets"].apply(calcular)

```

```python
def popularidad(fila):
    resultado = fila["followees"] / fila["followers"]
    return resultado

popu = df["popular"] = df.apply(popularidad, axis = 1)
```
